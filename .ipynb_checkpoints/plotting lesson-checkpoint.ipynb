## plotting lesson

```python
import numpy
```

```python
import os
distance_file=os.path.join('data','data','distance_data_headers.csv')
distances= numpy.genfromtxt(fname=distance_file, delimiter=',',dtype='unicode')
print(distances)
```

```python
headers=distances[0]
data=distances[1:]
data=data.astype(numpy.float)
```

```python
print(headers)
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,1]),data[:,1]
```

```python
sample=headers[1]
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance(angstroms)')
fig_1=plt.plot(data[:,0],data[:,1],label=sample)
plt.legend()

plt.savefig(F'{sample}.png')
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,0], data[:,1], label=headers[1])
plt.plot(data[:,0], data[:,2], label=headers[2])
plt.legend()
```

```python
plt.figure()
for col in range(1, len(headers)):
    fig=plt.plot(data[:,0],data[:,col], label=headers[col])
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()
```

```python
plt.figure()
for col in range(1, len(headers)):
    plt.figure()
    fig=plt.plot(data[:,0],data[:,col], label=headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png',dpi=300)
```
